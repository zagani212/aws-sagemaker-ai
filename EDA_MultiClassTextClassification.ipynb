# Load Data

```python
import pandas as pd

s3_path = 's3://news-aggregator-abdelhak-27-04-2026/newsCorpora.csv'
df = pd.read_csv(s3_path, sep='\t', names=['ID', 'TITLE', 'URL', 'PUBLISHER', 'CATEGORY', 'STORY', 'HOSTNAME', 'TIMESTAMP'])
```

```python
df.head()
```

```python
df_work = df.copy()
```

```python
df_work=df_work[['TITLE', 'CATEGORY']]
```

```python
df_work
```

```python
my_dict = {
    'e': 'Entertainment',
    'b': 'Business',
    't': 'Science',
    'm': 'Health'
}
```

```python
def update_category(x):
    return my_dict[x]

df_work['CATEGORY'] = df_work['CATEGORY'].apply(lambda x: update_category(x))
```
